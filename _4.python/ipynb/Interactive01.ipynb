```python

from ipywidgets import interactive
from IPython.display import display

def f(a, b):
    display(a + b)
    return a+b

# interactive é è¨­ä¸æœƒé¡¯ç¤º UI
w = interactive(f, a=10, b=20)

# é¡¯ç¤º UI
display(w)

# å–å¾—è¼¸å…¥å€¼
w.kwargs

# å–å¾—å‡½æ•¸å‚³å›žå€¼
w.result

```

```python
# è¨ˆç®—å¾ˆè¤‡é›œçš„å‡½æ•¸
# åŸ·è¡Œçµæžœï¼šæœƒå‡ºç¾ä¸€å€‹ã€ŽRun Interactã€æŒ‰éˆ•ï¼ŒæŒ‰ä¸‹å¾Œæ‰æœƒåŸ·è¡Œå‡½æ•¸

def slow_function(i):
    print(int(i),list(x for x in range(int(i)) if
                str(x)==str(x)[::-1] and
                str(x**2)==str(x**2)[::-1]))
    return

# æ‰‹å‹•æ›´æ–°
from ipywidgets import interact_manual, FloatSlider
interact_manual(slow_function,i=FloatSlider(min=1e5, max=1e7, step=1e5));


#ä¹Ÿå¯ä»¥ç›´æŽ¥åŠ åƒæ•¸ continuous_update=Falseã€‚
interactive(slow_function,i=FloatSlider(min=1e5, max=1e7, step=1e5, continuous_update=False));

```

```python
# ç•«é¢è¨­è¨ˆ(UI Layout)

import ipywidgets as widgets

# ä½¿ç”¨ Slider
a = widgets.IntSlider()
b = widgets.Text()
c = widgets.IntSlider()

# HBoxï¼šæ°´å¹³æŽ’åˆ—
ui = widgets.HBox([a, b, c])
def f(ID, Name, Age):
    print((ID, Name, Age))

out = widgets.interactive_output(f, {'ID': a, 'Name': b, 'Age': c})

display(ui, out)
```

```python
#äº’ç›¸é—œè¯çš„è¼¸å…¥æ¬„ä½
#ä¸€å€‹è¼¸å…¥æ¬„ä½å¯èƒ½æœƒå› å¦ä¸€è¼¸å…¥æ¬„ä½æœ‰æ‰€è®ŠåŒ–ï¼Œä¾‹å¦‚è¯èª¼ç”·æ€§é™40æ­²ã€å¥³æ€§é™30æ­²ï¼Œé¡žä¼¼æƒ…æ³ç¨‹å¼ç¢¼å¦‚ä¸‹ã€‚

x_widget = FloatSlider(min=0.0, max=10.0, step=0.05)
y_widget = FloatSlider(min=0.5, max=10.0, step=0.05, value=5.0)

def update_x_range(*args):
    # x è¼¸å…¥æ¬„æœ€å¤§å€¼éš¨ y_widget è€Œè®ŠåŒ–
    x_widget.max = 2.0 * y_widget.value

# x_widget éš¨ y_widget è€Œè®ŠåŒ–
y_widget.observe(update_x_range, 'value')

def printer(x, y):
    print(x, y)

interactive(printer,x=x_widget, y=y_widget);
```

```python
#æ‡‰ç”¨
#ä»¥ä¸‹ä»¥ç°¡å–®ç·šæ€§è¿´æ­¸ç¹ªåœ–ç‚ºä¾‹ï¼Œè¨­å®šæ–œçŽ‡(m)åŠæˆªè·(b)ï¼Œç•«å‡ºè¿´æ­¸ç·šã€‚

from ipywidgets import interactive
import matplotlib.pyplot as plt
import numpy as np

def f(m, b):
    plt.figure(2)
    x = np.linspace(-10, 10, num=1000)
    plt.plot(x, m * x + b)
    plt.ylim(-5, 5)
    plt.show()

interactive_plot = interactive(f, m=(-2.0, 2.0), b=(-3, 3, 0.5))
output = interactive_plot.children[-1]
output.layout.height = '350px'
interactive_plot
```
