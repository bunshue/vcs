# 假設檢定(Hypothesis Testing)之顯著水準

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

## 統計誤差範圍計算

```python
# 在顯著水準5%下，抽樣誤差 = 1.96 * 標準誤差(σ)
# 1.96 * σ = 0.03 (標準誤差)
# σ = (p * (1-p) / n) ** (1/2)
# ==> n = (p * (1-p)) * (σ ** 2)
n = 0.5 * 0.5 / ((0.03/1.96) ** 2)
print(n)
```

## 計算抽樣誤差
## [2020年中華民國總統選舉民意調查 - 維基百科，自由的百科全書](https://zh.wikipedia.org/zh-tw/2024年中華民國總統選舉民意調查)

```python
# 計算抽樣誤差(第一筆【好好聽文創傳媒股份有限公司】)
n=1086
std = 1.96 * ((0.5 * 0.5 / n) ** (1/2))
print(std)
```

```python
# 計算2000個有效樣本之抽樣誤差
n=2000
std = 1.96 * ((0.5 * 0.5 / n) ** (1/2))
print(std)
```

## 雙尾檢定

```python
# 灰色面積的機率
mean, std = 0, 1
# norm.cdf：常態分配的累積密度函數
Upper_bound = norm.cdf(mean + 1.96 * std, mean, std)
Lower_bound = norm.cdf(mean - 1.96 * std, mean, std)
print(f'{Lower_bound:.3f}, {Upper_bound:.3f}, {Upper_bound - Lower_bound:.3f}')
```

```python
mean, std = 0, 1
x = np.linspace(mean - 4 * std, mean + 4 * std, 100)
y = norm.pdf(x, mean, std)
plt.plot(x, y)

# show the hypothesized population mean
plt.axvline(mean+1.96*std, color='red', linestyle='dashed', linewidth=2)
plt.axvline(mean-1.96*std, color='red', linestyle='dashed', linewidth=2)

# 左邊拒絕域
a = np.linspace(mean - 4 * std, mean - 1.96 * std, 100)
b = norm.pdf(a, mean, std)
plt.fill_between(a, b, 0, color = 'lightgray') 

# 右邊拒絕域
a = np.linspace(mean + 1.96 * std, mean + 4 * std, 100)
b = norm.pdf(a, mean, std)
plt.fill_between(a, b, 0, color = 'lightgray') 

plt.ylim(0)
plt.show()
```

## 單尾檢定

```python
mean, std = 0, 1
# norm.cdf：常態分配的累積密度函數
Upper_bound = norm.cdf(mean + 1.645 * std, mean, std)
print(f'{Upper_bound:.3f}, {1 - Upper_bound:.3f}')
```

```python
mean, std = 0, 1
x = np.linspace(mean - 4 * std, mean + 4 * std, 100)
y = norm.pdf(x, mean, std)
plt.plot(x, y)

# show the hypothesized population mean
plt.axvline(mean+1.645*std, color='red', linestyle='dashed', linewidth=2)

# 右邊拒絕域
a = np.linspace(mean + 1.645 * std, mean + 4 * std, 100)
b = norm.pdf(a, mean, std)
plt.fill_between(a, b, 0, color = 'lightgray') 

plt.ylim(0)
plt.show()
```
