```python
import pandas as pd

products = {"channel": ["網路","網路","電視","電視","郵購","郵購"],
            "company": ["EHS","Momo","EHS","Viva","Momo","EHS"],
            "sales":   [11.22,23.50,12.99,15.95,25.75,11.55]}
oridnals = ["A","B","C","D","E","F"]
df = pd.DataFrame(products, index=oridnals)
df
```

```python
score = [4, 3, 5, 7, 5, 8]
df["score"] = score
df
```

```python
pivot_tb = df.pivot_table(index='channel',
                          columns='company',
                          values='sales')
pivot_tb
```

```python
pivot_tb = df.pivot_table(index='channel',
                          columns='company',
                          values='score')
pivot_tb
```
