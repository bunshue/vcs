```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
from scipy import special
```

```python
x = np.arange(5, 20)
y = special.exp2(x/3.0)
plt.plot(x, y, 'o')
plt.show()
```

```python
f = interpolate.interp1d(x, y)
x1 = np.arange(5, 20)
y1 = f(x1)
plt.plot(x, y, "o", x1, y1, "--")
plt.show()
```
