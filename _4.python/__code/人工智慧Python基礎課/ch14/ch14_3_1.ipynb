```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def f(x):
    return x**2 + 15*np.sin(x)
x = np.arange(-10, 10, 0.1)
plt.plot(x, f(x)) 
plt.show() 
```

```python
result = optimize.minimize(f, x0=0)
print(result.x)  
```

```python
plt.plot(x, f(x))
plt.plot(result.x, f(result.x), "o")
plt.show() 
```
