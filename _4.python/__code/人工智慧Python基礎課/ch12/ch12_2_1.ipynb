```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = [-1, -4.3, 15, 21, 31]
plt.plot(data) 
plt.show()
```

```python
plt.plot(data, "o--b")
plt.show()
```

```python
days = range(1, 7)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5]
plt.plot(days, celsius)
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, x, cosinus)
plt.show()
```
