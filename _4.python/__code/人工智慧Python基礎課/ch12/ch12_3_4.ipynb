```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
labels = ["Python","C++","Java","JS","C","C#"]
ratings = [5, 6, 15, 3, 12, 4]
```

```python
plt.pie(ratings, labels=labels)
plt.title("程式語言的使用率") 
plt.axis("equal")
plt.show()
```

```python
patches, texts = plt.pie(ratings, labels=labels)
plt.legend(patches, labels, loc="best")
plt.title("程式語言的使用率") 
plt.axis("equal")
plt.show()
```
