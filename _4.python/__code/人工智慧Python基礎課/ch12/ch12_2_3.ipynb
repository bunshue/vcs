```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = range(1, 7)
celsius = [25.6, 23.2, 18.5, 28.3, 26.5, 30.5]
plt.plot(days, celsius, "g--s")
plt.xlabel("日數")
plt.ylabel("攝氏溫度")
plt.show()
```

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o",
         x, cosinus, "g--")
plt.xlabel("徑度")
plt.ylabel("振幅")
plt.title("Sin 和 Cos 波形")
plt.show()
```
