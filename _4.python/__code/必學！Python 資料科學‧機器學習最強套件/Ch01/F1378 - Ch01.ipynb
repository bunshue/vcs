#1-2-2	 更新變數的值

```
x = 1 #建立變數名稱 x
print(x)

x = 5 #修改 x 的值
print(x)

x = x + 1 #修改 x 的值 (將 x 加 1)
print(x)

x += 1 #寫法等同於 x = x + 1
print(x)
```

#1-5-2	 if... else

```
n = 9
print('檢查數字...')
if n >= 10:
  print('數字超過門檻')
else:
  print('數字未達門檻')
```

#1-5-3	 if... elif... else...

```
n = 16
if n >= 20:
  print(n, '是很大的數字')
elif n >= 15:     #n >= 15 但< 20時
  print(n, '是中等的數字')
elif n >= 10:     #n >= 10 但< 15時
  print(n, '是普通的數字')
else:           #n < 10 時
  print(n, '是較小的數字')
```

#1-5-4	 and、or、not

```
x = 14
y = 28

print(x > 8 and y < 14)
print(x > 8 or y < 14)
print(x > 8 and not y < 14)
```

```
year = 2020

if (year % 4 == 0 and year % 100 != 0) or year % 400 == 0:
  print(year, '是閏年')
else:
  print(year, '是平年')
```
