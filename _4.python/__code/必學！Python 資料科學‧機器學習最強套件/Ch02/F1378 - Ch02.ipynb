#2-1-1	 建立 list

```
colors = ['red', 'blue', 'yellow']
print(colors)
print(type(colors))
```

#2-1-2	 在 list 放入不同型別的值或子 list

```
data = ['apple', 3, [4.5, 'car', True]]
print(data)
```

#2-1-3	 以索引從 list 取出元素

```
data = ['apple', 3, [4.5, 'car', True]]

print(data[0])
print(data[1])
print(data[2])
print(data[2][0])
print(data[2][2])
```

```
print(data[-1])
print(data[-2])
print(data[-3])
print(data[-1][-1])
```

#2-1-4	 用切片從 list 取出一群元素

```
c = ['a', 'b', 'c', 'd', 'e', 'f', 'g']

print(c[:])
print(c[:3])
print(c[3:])
print(c[2:5])
print(c[::2])
print(c[1:6:2])
```

#2-1-5	 在 list 新增元素

```
fruits = ['蘋果', '香蕉', '橘子']

fruits.append('葡萄')
print(fruits)

fruits += ['鳳梨', '草莓']
print(fruits)
```

#2-1-6	 從 list 刪除元素

```
fruits = ['蘋果', '香蕉', '橘子', '葡萄', '鳳梨', '草莓']

del fruits[0] #刪除索引 0 元素 (蘋果)
print(fruits)

fruits.remove('橘子')
print(fruits)
```

#2-2-1	 建立 dict

```
capitals = {'瑞士': '伯恩', '日本': '東京', '美國': '華盛頓特區'}
print(capitals)
```

```
language = {'name': 'Python',
        'version': 3.9,
        'types': ['int', 'float', 'str', 'bool']}

print(language)
```

#2-2-2	 以鍵從 dict 取出元素

```
capitals = {'瑞士': '伯恩', '日本': '東京', '美國': '華盛頓特區'}

print(capitals['美國'])
print(capitals['日本'])
```

#2-2-3	 在 dict 加入新元素

```
capitals = {'瑞士': '伯恩', '日本': '東京', '美國': '華盛頓特區'}
capitals['越南'] = '河內'
print(capitals)

capitals.update({'英國': '倫敦', '法國': '巴黎'})
print(capitals)
```

#2-2-4	 從 dict 刪除元素

```
capitals = {'瑞士': '伯恩', '日本': '東京', '美國': '華盛頓特區'}

del capitals['美國'] 
print(capitals)
```

#2-3-1	 基本 while 迴圈

```
n = 0
while n < 5:
  print(n)
  n += 1
```

#2-3-2	 使用 break 來脫離 while 迴圈

```
n = 0
while True:
  print(n)
  n += 1
  if n >= 5:
    break
```

#2-3-3	 用 continue 跳過後面的程式碼

```
n = 1
while n < 10:
  n += 1 #每次 n 遞增 1
  if n % 3 == 0:
    continue #若 n 是 3 的倍數就跳過下面的程式碼(不印)
  print(n)
```

#2-4-1	 基本 for 迴圈

```
animal_list = ['dog', 'cat', 'monkey', 'bird', 'elephant']

for animal in animal_list:
  print(animal)
```

#2-4-2	 走訪二維 list

```
fruits = [
  ['apple', 'red'],
  ['banana', 'yellow'],
  ['guava', 'green'],
  ]
for fruit in fruits:   #第一層 for 迴圈 (走訪 list)
  for item in fruit:    #第二層 for 迴圈 (走訪子 list)
    print(item)
```

#2-5-1	 在 for 迴圈搭配 range() 作為索引

```
print(list(range(5)))
```

```
animal_list = ['dog', 'cat', 'monkey', 'bird', 'elephant']
for index in range(5):
  print('item', index, '=', animal_list[index])
```

#2-5-2	 在 for 迴圈搭配 enumerate() 來同時走訪索引及元素

```
print(list(enumerate(animal_list)))
```

```
animal_list = ['dog', 'cat', 'monkey', 'bird', 'elephant']
for index, item in enumerate(animal_list):
  print('item', index, '=', item)
```

#2-5-3	 用 zip() 同時走訪多個 list

```
index_list = ['a', 'b', 'c', 'd', 'e']
animal_list = ['dog', 'cat', 'monkey', 'bird', 'elephant']
print(list(zip(index_list, animal_list)))
```

```
fruits = ['apple', 'peach', 'banana', 'guava', 'papaya']
colors = ['red', 'pink', 'yellow', 'green', 'orange']

for name, color in zip(fruits, colors):
  print(name, 'is', color)
```

#2-5-4	 用 for 走訪 dict

```
fruits = {
'apple': 'red',
'peach': 'pink',
'banana': 'yellow',
'guava': 'green',
'papaya': 'orange'
}
print(list(fruits.items()))
```

```
fruits = {
'apple': 'red',
'peach': 'pink',
'banana': 'yellow',
'guava': 'green',
'papaya': 'orange'
}

for name, color in fruits.items():
  print(name, 'is', color)
```
