#3-1-2	 用 len() 取得容器長度

```
characters = ['a', 'b', 'c', 'd', 'e', 'f', 'g']
text = 'Hello World!'

print(len(characters))
print(len(text))
```

#3-1-3	 list 排序：sorted() 與 list.sort()

```
numbers = [5, 3, 2, 6, 7, 4, 1]

print(sorted(numbers)) #用 sorted() 不會改變原 list
print(numbers)

numbers.sort()    #用 list.sort() 會改變原 list
print(numbers)
```

#3-1-4	 反轉容器：reversed() 和 list.reverse()

```
numbers = [1, 2, 3, 4, 5, 6, 7]

print(list(reversed(numbers)))
print(numbers)

numbers.reverse()
print(numbers)
```

#3-1-5	 用 count() 統計值出現的數量

```
numbers = [2, 3, 1, 5, 4, 5, 4, 1, 5, 1]
print(numbers.count(5)) #印出傳回 numbers 中 5 出現的次數

text = 'banana'
print(text.count('a'))
```

#3-1-6	 用 index() 尋找值的索引

```
numbers = [2, 3, 1, 5, 4, 5, 4, 1, 5, 1]
print(numbers.index(5)) #數字 5 在 numbers 最先出現的索引位置

text = 'banana'
print(text.index('n'))
```

#3-2-1	 定義函式

```
def sing():
  print('唱首歌吧!') 

sing()
```

#3-2-2	 給函式加入參數

```
def sing(name):
  print('{}, 唱首歌吧!'.format(name))

sing('Daisy')
```

```
def sing(name, verb):
  print('{}, 請{}!'.format(name, verb))

sing('Daisy', '唱首歌')
```

#3-2-3	 讓函式傳回值

```
def mod(a, b):
  return a % b #傳回參數 a 除以 b 的餘數

print(mod(20, 7))
```

```
def mod(a, b):
  q = a // b
  r = a % b
  return q, r #傳回兩個值, 分別是商跟餘數
x, y = mod(20, 7) #接收兩個傳回值, 分別給 x, y
print('商數={}, 餘數={}'.format(x, y))
```

#3-2-4	 給函式參數指定預設值

```
def sing(name='無名氏', verb='發呆'): #參數與其預設值
  print('{}, 請{}!'.format(name, verb))


sing('Daisy', '唱首歌')
sing(verb='演說', name='Daisy')
sing('Daisy')
sing(verb='尖叫')
sing()
```

#3-3-3	 自行定義一個類別

```
class MyProduct:
  def __init__(self, n, p): #定義 __init__ 函式
    self.name = n
    self.price = p


book = MyProduct('書', 700)
print(book)
print(book.name)
print(book.price)
```

#3-3-4	 替物件加入 method

```
class MyProduct:
  def __init__(self, n, p):
    self.name = n
    self.price = p
  def summary(self): #自訂 method
    print('商品名稱: {}\n價格: {}'.format(self.name, self.price))

  def discount(self, rate): #自訂 method
    self.price *= rate

book = MyProduct('書', 700)

book.discount(0.79)
book.summary()
```

```
car = MyProduct('汽車', 700000)
car.discount(0.9)
car.summary()
```

#3-4-1	 datetime 物件

```
import datetime as dt

x = dt.datetime(2020, 10, 22)
print(x)

x = dt.datetime(year=2020, month=10, day=22)
print(x)

y = dt.datetime(2020, 10, 22, 10, 30, 45) #設定日期與時間
print(y)
```

#3-4-2	 timedelta 物件

```
x = dt.timedelta(hours=1, minutes=30) #1 小時又 30 分
print(x)

y = dt.timedelta(days=1, seconds=30) #1 天又 30 秒
print(y)
```

#3-4-3	 用 timedelta 來增減 datetime 或 timedelta 的時間

```
import datetime as dt
x = dt.datetime(2020, 10, 22, 10, 30, 45) #原始時間
y = dt.timedelta(days=1, hours=2, minutes=5)

print(x)
print(x + y) #用 timedelta 來增減 datetime 的時間
print(x - y)
print(x + y * 2)
```

#3-4-4	 將 datetime 時間以格式化方式輸出

```
import datetime as dt
x = dt.datetime(2020, 10, 22, 10, 30, 45)

s1 = x.strftime('%Y/%m/%d %H-%M-%S')
print(s1)

s2 = x.strftime('%Y 年 %m 月 %d 日 %H : %M : %S')
print(s2)
```

#3-4-5	 用字串來建立 datetime 物件

```
import datetime as dt

s = '2020/10/22 10-30-45' #含有特定格式之日期時間字串
x = dt.datetime.strptime(s, '%Y/%m/%d %H-%M-%S')

print(x)
print(type(x))
```
