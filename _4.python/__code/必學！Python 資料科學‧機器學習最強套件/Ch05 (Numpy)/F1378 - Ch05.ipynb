#5-1 體驗 NumPy 的高速運算

```
import numpy as np
from numpy.random import rand
import time


N = 150
matA = np.array(rand(N, N))
matB = np.array(rand(N, N))
matC = np.array([[0] * N for _ in range(N)])


# 使用 Python 計算
start = time.time() 
for i in range(N):
    for j in range(N):
        for k in range(N):
            matC[i][j] = matA[i][k] * matB[k][j]
print("Python 的計算結果：%f[sec]" % float(time.time() - start))

# 使用 NumPy 計算

start = time.time()
matC = np.dot(matA, matB)


# 如果NumPy出現0.00[sec], 不代表真的是0, 而是小數點第2位後不表示的結果
print("NumPy 的計算結果：%f[sec]" % float(time.time() - start))
```

#5-2-4	 陣列的四則計算

```
# 使用 Python 的 list
storages = [1, 2, 3, 4]
new_storages = []
for n in storages:
  n += n
  new_storages.append(n)
print(new_storages)
```

```
# 使用 NumPy 陣列
import numpy as np
storages = np.array([1, 2, 3, 4])
storages += storages
print(storages)
```

```
import numpy as np
arr_1 = np.array([2, 4, 6, 8, 10])
arr_2 = np.array([1, 3, 5, 7, 9])
# arr + arr (相加)
print('arr_1 + arr_2:')
print(arr_1 + arr_2)
print('--------------------------')
# arr - arr (相減)
print('arr_1 - arr_2:')
print(arr_1 - arr_2)
print('--------------------------')
# arr ** 3 (三次方)
print('arr_1 ** 3:')
print(arr_1 ** 3)
print('--------------------------')
# arr_1 / arr_2(相除)
print('arr_1 / arr_2:')
print(arr_1 / arr_2)
print()
```

#5-2-5	 體驗好用的 NumPy 函式

```
import numpy as np
arr = np.array([4, -9, 16, -4, 20])
print(arr)
arr_abs = np.abs(arr)
print('絕對值:',arr_abs)
print('e為底數:',np.exp(arr_abs))
print('平方根:',np.sqrt(arr_abs))
```

```
import numpy as np
arr1 = np.array([2, 5, 7, 9, 5, 2])
arr2 = np.array([2, 5, 8, 3, 1])
new_arr1 = np.unique(arr1)
print('剔除arr1重複元素:',new_arr1)
print('聯集:',np.union1d(new_arr1, arr2))
print('交集:',np.intersect1d(new_arr1, arr2))
print('差集:',np.setdiff1d(new_arr1, arr2))
```

#5-3-5	 陣列排序

```
import numpy as np
arr = np.array([[8, 4, 2],
[3, 5, 1]])
print('---------------原陣列----------------')
print(arr)
print('----------對 arr 以軸 1 方向排序---------')
print(np.sort(arr))
print('----------對 arr 以軸 0 方向排序---------')
arr.sort(axis=0)
print(arr)
```

```
print('-------------argsort排序-------------')
print(arr.argsort())
```
