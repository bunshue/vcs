#11-2-6 圖片旋轉

```
img = cv2.imread(r'/content/drive/MyDrive/Colab_Notebooks/F1378/Ch11_(OpenCV)/photo/sample.jpg')  #請依11-2-1節的說明，換成您自己的雲端硬碟路徑
```

```
aff_matrix = cv2.getRotationMatrix2D((img.shape[1]/2, img.shape[0]/2), 30, 0.8)
```

```
img_rotate = cv2.warpAffine(img, aff_matrix, (img.shape[1], img.shape[0]))
```

```
cv2.imshow('Sample pic', img_rotate)
```

#11-3-2 遮罩

```
img = cv2.imread(r'/content/drive/MyDrive/Colab_Notebooks/F1378/Ch11_(OpenCV)/photo/sample.jpg')   #請依11-2-1節的說明，換成您自己的雲端硬碟路徑
mask = cv2.imread(r'/content/drive/MyDrive/Colab_Notebooks/F1378/Ch11_(OpenCV)/photo/mask.jpg',cv2.IMREAD_GRAYSCALE)   #請依11-2-1節的說明，換成您自己的雲端硬碟路徑
```

```
mask = cv2.resize(mask, (img.shape[1], img.shape[0]))
img_masked = cv2.bitwise_and(img, img, mask=mask)
cv2.imshow('Sample pic', img_masked)
```

```
mask = cv2.bitwise_not(mask)		
img_masked = cv2.bitwise_and(img, img, mask=mask)	
cv2.imshow('Sample pic', img_masked)
```
