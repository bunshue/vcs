```python
class TVSet:
    # 建構式
    def __init__(self):
        self._power = False
        self._channel = 1
        self._volume = 0
        
    @property
    def power(self):
        return self._power
    
    # 切換電源狀態
    def switchPower(self):
        self._power = not self._power
        
    @property
    def channel(self):
        return self._channel
    
    # 直接設定頻道
    @channel.setter
    def channel(self, v):
        # 檢查邊界狀況
        if v>=1 and v<=255:
            self._channel = v
    
    # 往後一個頻道
    def channelNext(self):
        # 檢查邊界狀況，確認是否要循環
        if self._channel==255:
            self._channel = 0
        else:
            self._channel += 1
    
    # 前一個頻道
    def channelPrev(self):
        # 檢查邊界狀況，確認是否要循環
        if self._channel==0:
            self._channel = 255
        else:
            self._channel -= 1
            
    @property
    def volume(self):
        return self._volume
    
    def volumeUp(self):
        # 檢查邊界狀況
        if self._volume<10:
            self._volume += 1
            
    def volumeDown(self):
        # 檢查邊界狀況
        if self._volume>0:
            self._volume -= 1
            
tv1 = TVSet()
# 打開電源
tv1.switchPower()
# 使用channel的setter
tv1.channel = 64
# 音量無法直接設定，要一步步轉大聲
for i in range(5):
    tv1.volumeUp()

tv2 = TVSet()
# 打開電源
tv2.switchPower()
# 切到前一台，其實會變成channel 255
tv2.channelPrev()
# 大聲一格，音量會變成1
tv2.volumeUp()
```
