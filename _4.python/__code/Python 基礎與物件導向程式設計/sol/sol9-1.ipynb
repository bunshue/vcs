```python
from threading import Thread
import time

# 模擬兔子賽跑的狀況
def rabbitRun():
    progress = 0
    while progress<30:
        progress += 1
        print('兔子跑了', progress, '公尺')
        time.sleep(1)
        if progress%7==0:
            time.sleep(10)
    print('兔子到達終點！')

# 模擬烏龜賽跑的狀況
def turtleRun():
    progress = 0
    while progress<30:
        progress += 0.5
        print('烏龜跑了', progress, '公尺')
        time.sleep(1)
    print('烏龜到達終點！')

thr1 = Thread(target=rabbitRun)
thr2 = Thread(target=turtleRun)
# 龜兔賽跑開始！
thr1.start()
thr2.start()
```
