### 操作資料框架

程式4.14　以pandas的資料框架格式載入範例資料

```python
titanic = sns.load_dataset("titanic")
titanic
```

程式4.15　取得第一列的元素的範例

```python
titanic.iloc[0]
```

程式4.16　取得一個欄位的範例

```python
titanic_class = titanic["class"]
titanic_class
```

程式4.17　取得一個欄位的範例

```python
len(titanic)
```

程式4.18　利用describe函數確認資料概要的範例

```python
titanic.describe()
```

程式4.19　計算欄位的資料筆數的範例

```python
titanic_class = titanic["class"].value_counts()
titanic_class
```

程式4.20　計算欄位特殊元素的範例

```python
titanic_unique = titanic["class"].nunique()
titanic_unique
```

程式4.21　摘要資料再輸出資料筆數

```python
titanic_sex_class = titanic.groupby("sex")["class"].value_counts()
titanic_sex_class
```

程式4.22　計算每個特定變數的平均值

```python
titanic_group_mean = titanic.groupby("sex").mean()
titanic_group_mean
```

程式4.23　計算每種性別的平均值

```python
titanic_group_mean = titanic.groupby("sex", as_index=False).mean()
titanic_group_mean
```

程式4.24　以兩個變數摘要再計算平均值的範例

```python
titanic_group_mean2 = titanic.groupby(["sex", "class"], as_index=False).mean()
titanic_group_mean2
```

程式4.25　摘要資料筆數的範例

```python
cross_class = pd.crosstab(titanic["who"], titanic["class"])
cross_class
```

程式4.26　正規化範例

```python
cross_nmrl = pd.crosstab(titanic["who"], titanic["class"], normalize="index")
cross_nmrl
```

程式4.27　正規化範例

```python
titanic_female = titanic[titanic["sex"] == "female"]
titanic_female
```

程式4.28　取得符合條件的欄位

```python
titanic_female = titanic.query("sex == 'female'")
titanic_female
```

程式4.29　排序資料的範例

```python
titanic_female_sort = titanic_female.sort_values("fare")
titanic_female_sort
```

程式4.30　降冪排序的範例

```python
titanic_female_sort = titanic_female.sort_values("fare", ascending=False)
titanic_female_sort
```

程式4.31　變更欄位名稱的範例

```python
# 將age這個欄位變更為年齡
titanic_rename = titanic_female_sort.rename(columns={"age": "年齡"})
titanic_rename
```
