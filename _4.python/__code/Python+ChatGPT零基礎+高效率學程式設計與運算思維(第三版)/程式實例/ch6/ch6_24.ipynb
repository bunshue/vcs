```python
# ch6_24.ipynb
cars = ['honda','bmw','toyota','ford']     
print(f"目前串列內容 = {cars}")
print("使用sort()由小排到大")
cars.sort()            
print(f"排序串列結果 = {cars}")
print("使用sort()由大排到小")
cars.sort(reverse=True)            
print(f"排序串列結果 = {cars}")
print("="*60)
nums = [5, 3, 9, 2]
print("目前串列內容 = ",nums)
print("使用sort()由小排到大")
nums.sort()            
print(f"排序串列結果 = {nums}")
print("使用sort()由大排到小")
nums.sort(reverse=True)            
print(f"排序串列結果 = {nums}")
```
