# 國內主要金融指標

資料來源：

* [政府開放資料平台 - 國內主要金融指標](https://data.gov.tw/dataset/30815)

```python
import requests
import pandas as pd
from io import StringIO
```

```python
url_by_year = "https://apiservice.mol.gov.tw/OdService/download/A17030000J-000037-51i"
url_by_month = "https://apiservice.mol.gov.tw/OdService/download/A17030000J-000037-3zL"
```

```python
r = requests.get(url_by_year)
```

```python
df = pd.read_csv(StringIO(r.text))
```

```python
df.set_index('年度')
```

```python
r = requests.get(url_by_month)
df = pd.read_csv(StringIO(r.text))
```
