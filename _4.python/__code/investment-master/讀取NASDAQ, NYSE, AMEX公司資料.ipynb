# 讀取 NASDAQ, NYSE, AMEX 公司資料

參考資料：

* [Company List (NASDAQ, NYSE, AMEX](https://www.nasdaq.com/screening/company-list.aspx)

```python
import pandas as pd
```

```python
url_template = "https://www.nasdaq.com/screening/companies-by-industry.aspx?exchange={}&render=download"
```

```python
url = url_template.format('NASDAQ')
```

```python
df = pd.read_csv(url)
```

```python
df = df.loc[:, ~df.columns.str.contains("Unnamed")]
```
