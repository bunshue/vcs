# 上市公司基本資料

```python
import requests
import pandas as pd
from io import StringIO
```

```python
# 讀取上市公司基本資料
r = requests.get("http://dts.twse.com.tw/opendata/t187ap03_L.csv")
r.encoding = "big5"
df = pd.read_csv(StringIO(r.text), index_col=False, skiprows=1)
df.drop(df.index[len(df.index)-1], inplace=True)
df
```
