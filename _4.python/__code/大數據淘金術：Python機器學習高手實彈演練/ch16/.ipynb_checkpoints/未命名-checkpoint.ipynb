```python
import pandas as pd

df = pd.read_csv('weibo_train_data.txt',sep='\t',header=None)
df = df.rename(columns={0:'uid',1:'m',2:'datetime',3:'f',4:'c',5:'l'})
print(df.shape)
```

```python
print(df[5:10])
df[6:10].to_csv('weibo.csv',index=False, encoding='gbk',header=None)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

counts = df[3].value_counts()
print(type(counts))
plt.plot(counts)
```

```python
df[3].value_counts()
print(len(df[df[3] == 0])/len(df))
print(len(df[df[3] < 50])/len(df))
```

```python
grp = df.groupby(0)
print(len(grp))
```

```python
df['feed_back'] = df[3]+df[4]+df[5]
```

```python
counts = grp['feed_back'].count()
#plt.hist(counts, 200)

means = grp['feed_back'].mean()
count = 0
for i in means:
    print(i)
    if i == 0:
        count += 1
print(count)
```
