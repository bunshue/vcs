### 5.4.1  從網絡接口讀取數據

```python
import urllib.request

url = 'http://hq.sinajs.cn/' # 接口
values={'list':'sh601688'}   # 參數
data=urllib.parse.urlencode(values)
new_url=url+"?"+data
req = urllib.request.Request(new_url)  
html = urllib.request.urlopen(req).read()
print(html.decode("gb2312")) # 返回字符集爲GB2312,轉碼後顯示
```

### 5.4.2  抓取網站數據

```python
import urllib.request
from bs4 import BeautifulSoup

response = urllib.request.urlopen("https://blog.csdn.net/xieyan0811")
html = response.read().decode("utf-8","ignore") # 返回網頁爲utf-8編碼，解碼時忽略錯誤
reg = r'http://'  
soup = BeautifulSoup(html, 'html.parser')
for link in soup.find_all('a'):
    addr = link.get('href')
    # 顯示包含關鍵字的所有地址
    if addr != None and addr.find('xieyan0811/article') != -1: 
        print(addr)
```

### 5.4.3  使用POST方式抓取數據

```python
import requests
params = {'key1': 'value1', 'key2': 'value2'}
r = requests.post("http://httpbin.org/post", data=params)
print(r.text)
```

### 5.4.4  轉換HTML文件

```python
# 新版本的python中，已經不支持setdefaultencoding函數了，請在Python 2中執行

import sys
stdi, stdo, stde = sys.stdin, sys.stdout, sys.stderr
reload(sys)
sys.setdefaultencoding('utf8') # 設置字符集
sys.stdin, sys.stdout, sys.stderr = stdi, stdo, stde # 重置標準輸入輸出
```

```python
import html2text
import urllib.request 

html=urllib.request.urlopen('https://blog.csdn.net/xieyan0811').read() # 讀取文件內容
print(html2text.html2text(str(html)))

h = html2text.HTML2Text()
h.ignore_links=True # 忽略超鏈接
print(h.handle(str(html)))
```
