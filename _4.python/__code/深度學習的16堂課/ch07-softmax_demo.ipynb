### 7.4.2 softmax 函數的運算方式

```python
from math import exp
```

```python
z = [-1.0, 1.0, 5.0]
```

```python
exp(z[0])
```

```python
exp(z[1])
```

```python
exp(z[2])
```

```python
total = exp(z[0]) + exp(z[1]) + exp(z[2])
```

```python
total
```

```python
exp(z[0])/total
```

```python
exp(z[1])/total
```

```python
exp(z[2])/total
```
