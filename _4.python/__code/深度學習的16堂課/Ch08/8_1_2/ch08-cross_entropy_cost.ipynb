###8.1.2 交叉熵 (Cross Entropy) 損失函數

```python
from numpy import log
```

```python
def cross_entropy(y, yhat):
    return -1*(y*log(yhat) + (1-y)*log(1-yhat))
```

```python
cross_entropy(1, 0.9997)
```

```python
cross_entropy(1, 0.9)
```

```python
cross_entropy(1, 0.6)
```

```python
cross_entropy(1, 0.1192)
```

```python
cross_entropy(0, 0.1192)
```

```python
cross_entropy(1, 1-0.1192)
```
