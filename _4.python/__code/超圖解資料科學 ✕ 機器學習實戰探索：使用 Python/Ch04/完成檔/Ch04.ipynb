# 4-2-8 新增列資料

# 實作-新增列資料【EX4-2.8.ipynb】 

# Step 01

```
from google.colab import drive
drive.mount('/content/MyGoogleDrive')
import pandas as pd
df=pd.read_csv(i_filepath + '學生成績檔-4-2.8.csv')
df.tail()
```

# Step 02

```
s = ['1080041', '張三', '1', '1080041@sun.tc.edu.tw',90, 85, 90, 95, 95]
df1 = pd.DataFrame(data=[s], columns=df.columns)
df = df.append(df1, ignore_index=True)
df.tail()
```

# Step 03

```
c = ['第1次平時考','第2次平時考','第3次平時考','第4次平時考','第5次平時考']
df['總分'] = df[c].sum(axis='columns')
df['平均'] = df[c].mean(axis='columns')
df = df.sort_values('平均',ascending=False)
df['排名'] = list(range(1,len(df)+1)) 
df['ID'] = df.ID.astype(str)
df = df.sort_values(['ID'], ascending=True)
df.head()
```

# Step 04 儲存結果【學生成績檔-4-2.8-ANS.csv】

```
filename='學生成績檔-4-2.8-ANS.csv'       
df.to_csv(o_filepath + filename, index=False)
```
