#實作-條件判斷(多向)【EX2-2.7c.ipynb】

# Step 01

```
h = float(input('身高(m):'))   
w = float(input('體重(Kg):'))   
BMI = w / (h**2)
print('BMI:',BMI)             
if BMI < 18.5:
  print('過輕')
elif BMI >=24:
  print('過重')
else:
  print('合適')
```

# 2-2-8  重複執行

#實作-重複執行(單層for)【EX2-2.8a.ipynb】

# Step 01

```
寶可夢 = ['妙蛙種子', '妙娃草', '噴火龍', '綠毛蟲']
for 抓寶 in 寶可夢:
    print(抓寶)
```

#實作-重複執行(單層for)【EX2-2.8b.ipynb】

# Step 01

```
t = ['國文','英文','數學','資訊科技']
s = [84,92,88,95]
for i in [0,1,2,3]:
  print(t[i])
for i in range(4):
  print(s[i])
```

# 實作-重複執行(雙層for)【EX2-2.8c.ipynb】

# Step 01

```
cs = [['國文','英文','數學','資訊科技'],[84,92,88,95]]
for i in [0,1]:
  for j in [0,1,2,3]:
    print(cs[i][j])
```

# 加廣知識：雙層迴圈改成單層迴層

```
cs = [['國文','英文','數學','資訊科技'],[84,92,88,95]]
for i in [0,1,2,3]:
    print(cs[0][i],cs[1][i])
```

#實作-重複執行(for 求總和)【EX2-2.8d.ipynb】

#實作-終極密碼戰【EX2-2.9b.ipynb】

# Step 01

```
import random as rd
x = rd.randint(1,10)
for i in range(3):              
  s = int(input('猜數字1-10:'))
  if x == s:
    print('猜中')
    break
  elif x > s:
    print('大一點')
  elif x < s:
    print('小一點')
print('遊戲結束!')
```
