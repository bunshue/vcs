我們來做個標準常態分佈的!!

```python
x = np.random.randn(500)
y = np.random.randn(500)
```

```python
plt.scatter(x,y)
```

試著做三群的數據。

```python
cx0 = 0
cy0 = 0

cx1 = -3
cy1 = 3

cx2 = 3
cy2 = 3
```

每一群都是 500 個點。

```python
x0 = np.random.randn(500) + cx0
y0 = np.random.randn(500) + cy0

x1 = np.random.randn(500) + cx1
y1 = np.random.randn(500) + cy1

x2 = np.random.randn(500) + cx2
y2 = np.random.randn(500) + cy2
```

合併三群的點到 x, y 之中。

```python
x = np.concatenate((x0, x1, x2))
y = np.concatenate((y0, y1, y2))
```

現在第一群的點是第 0 類, 第二群是第 1 類, 第三群是第 2 類，所以我們要做個像這樣

    [0, 0, ..., 0, 1, 1, ..., 1, 2, 2, ..., 2]

的標記。

```python
c = np.repeat([0,1,2], 500)
```

```python
plt.scatter(x, y, c=c, cmap='Set1')
```

注意前面我們發現, 我們的數據有些問題。可能有些日子疫情沒有更新, 所以新增是 0, 結果加到下一天去。

在實際的工作中, 我們會花很多時間去調校, 做「資料清理」。

現在我們準備看看各國疫情高峰期, 走勢大概是怎麼樣的。

```python
de = dfm['Germany']
jp = dfm['Japan']
kr = dfm['Korea, South']
sg = dfm['Singapore']
tw = dfm['Taiwan*']
```

試以德國為例, 我們看看最高峰前後兩個月的樣子。

```python
n = de.argmax()
period = de[n-60:n+60]
ma = period.rolling(window=7).mean()[6:]
period[6:].plot()
ma.plot()
```

把剛剛的想法寫成一個函式, 簡簡單單就應用到不同國家。

```python
def peak_period(country):
    n = country.argmax()
    period = country[n-60:n+60]
    ma = period.rolling(window=7).mean()[6:]
    period[6:].plot()
    ma.plot()
```

我們再來看日本的情況。

```python
peak_period(jp)
```

可以發現日本大概一個月左右就到最高峰, 然後緩慢趨緩。

再來看看韓國的情形。

```python
peak_period(kr)
```

可以看到韓國大概花了兩個月到最高峰 (而且是相當高), 但兩個月後就差不多趨緩。

## 7. 用線性迴歸預測加州房價

現在我們用線性迴歸來預測加州的房價。

```python
from sklearn.datasets import fetch_california_housing
```

```python
housing = fetch_california_housing()
```

可以看看說明, 一共有 8 個特徵 (features), 20,640 筆數據。

```python
print(housing['DESCR'])
```

單獨把特徵的名稱找出來。

```python
housing.feature_names
```

特徵的數據是在 `.data` 中, 現在來建一個 Data Frame。

```python
cal = pd.DataFrame(housing.data, 
                   columns=housing.feature_names)
```

```python
cal.head()
```

把要預測的房價也放進來。

```python
cal['MEDV'] = housing.target
```

```python
cal.head()
```

可以用 `seaborn` 畫個美美的房價分佈圖。

```python
sns.distplot(cal.MEDV, bins=30)
```

再來可以做一個相關係數矩陣 (Correlation Matrix), 並且畫出 "heat map"。

```python
correlation_matrix = cal.corr().round(2)
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})

sns.heatmap(correlation_matrix, annot=True)
```

接下來我們把輸入輸出分好。

```python
X = cal.loc[:,"MedInc":"Longitude"].values
Y = cal.MEDV
```

切一下訓練及測試資料。

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                   test_size=0.2,
                                                   random_state=9487)
```

三部曲打造函數學習機!

```python
from sklearn.linear_model import LinearRegression
```

#### 第一部曲: 開一台函數學習機

```python
model = LinearRegression()
```

#### 第二部曲: 訓練

```python
model.fit(x_train, y_train)
```

#### 第三部曲: 測試

```python
y_predict = model.predict(x_test)
```

現在來計算我們預測的成績。

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
mse = mean_squared_error(y_test, y_predict)
r2 = r2_score(y_test, y_predict)

print("MSE =", mse)
print("R2 =", r2)
```

用很有創意的方法, 把圖給畫出來。

```python
plt.scatter(y_test, y_predict)
plt.xlim(0, 5.5)
plt.ylim(0, 5.5)
plt.plot([0,5.5],[0,5.5],'r')
```

## 8. 標準的分類問題: 這是哪種鳶尾花

我們現在來看看, 可不可以讓電腦辨識, 這是哪個亞種的鳶尾花?

```python
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

準備輸入及輸出數據, 注意 4 個特徵我們只用了兩個。

```python
x = iris.data
y = iris.target
```

```python
X = x[:, 2:]
Y = y
```

切分訓練及測試資料。

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y,
                                                   test_size=0.2,
                                                   random_state=0)
```

看一下整筆數據的分佈。

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='Paired')
```

再一次, 三部曲打造函數學習機。

#### 第一部曲：打開一台函數學習機

```python
from sklearn.svm import SVC
```

```python
clf = SVC()
```

#### 第二部曲：訓練

```python
clf.fit(x_train, y_train)
```

#### 第三部曲：測試

```python
y_predict= clf.predict(x_test)
```

看看我們模型預測和真實狀況差多少?

```python
y_predict - y_test
```

在測試資料中是全對!! 我們畫圖來看看整體表現如何?

```python
x0 = np.linspace(0, 7.5, 500)
y0 = np.linspace(0, 2.7, 500)

xm, ym = np.meshgrid(x0, y0)
P = np.c_[xm.ravel(), ym.ravel()]
z = clf.predict(P)
Z = z.reshape(xm.shape)
plt.contourf(xm, ym, Z, alpha=0.3)

plt.scatter(X[:,0], X[:,1], c=Y)
```
