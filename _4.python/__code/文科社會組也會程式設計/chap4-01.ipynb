```python
#chap4-01a
#i的初始值為1，終止值小於6，遞增值為1
for i in range(1,6,1):
  print(i, end='')
print()
#遞增值為1可以省略不寫
for i in range(1,6):
  print(i, end='')
print()
#初始值為0可以省略不寫
for i in range(6):
  print(i, end='')
```

```python
#chap4-01b
#i的初始值為5，終止值大於0，遞減值為-1
for i in range(5,0,-1):
  print(i, end='')
```

```python
#chap4-01c
#i的初始值為0，終止值小於6，遞增值為2
for i in range(0,6,2):
  print(i, end='')
```
