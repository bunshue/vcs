```python
#chap5-03a
def morning():
  print("早安")
def afternoon():
  print("午安")
def night():
  print("晚安")
def greetings(time):
  if time==1:
    print("早安")
  elif time==2:
    print("午安")  
  elif time==3:
    print("晚安")  
morning()
afternoon()
night()
greetings(1)
greetings(2)
greetings(3)
```

```python
#chap5-03b
def max1(a,b):
  if a>b:
    print(a)
  else:
    print(b)
def max2(a,b):
  if a>b:
    return a
  else:
    return b  
#沒有回傳值，就只有輸出結果
max1(3,5)
#有回傳值，可以再將傳回結果再作運算
a=max2(5,6)
b=max2(3,a)
print(b)
#上面三行可以直接寫成下面一行
print(max2(3,max2(5,6)))
```

```python
#chap5-03d
def test(): 
  global a
  global b
  print(a+b)
  a=3
  b=2
a=4
b=5
test()
print(a)
print(b)
```
