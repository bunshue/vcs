```python
#chap3-01a
#邏輯運算 not and or 應用
#相反
print("==Not==")
print(not True)
print(not False)
#或, 只有都是False才是False
print("==Or==")
print(False or False)
print(True or False)
print(False or True)
print(True or True)
#且, 只有都是True才是True
print("==And==")
print(False and False)
print(True and False)
print(False and True)
print(True and True)
print("==Test1==")
x=40
#判斷50<=x<=60
print(x>=50 and x<=60)
#迪摩根50<=x<=60 等價於 x<50 or x>60 以外
print(not (x<50 or x>60))
print(50<=x<=60)
print("==Test2==")
x=55
print(x>=50 and x<=60)
print(not (x<50 or x>60))
print(50<=x<=60)
print("==Test3==")
x=65
print(x>=50 and x<=60)
print(not (x<50 or x>60))
print(50<=x<=60)
```

```python
#chap3-01b
#比較運算
print("==Test1==")
#小於
a=5<3
print(a) #False
#大於
a=5>3
print(a) #True
#等於
a=5==3
print(a) #False
#不等於
a=5!=3
print(a) #True
#in 用來判斷有沒有出現在串列list
print("==Test2==")
x=1
y=[1,2,3,4]
print(x in y) #True
print(x not in y) #Falase
print(5 in y) #False
print(5 not in y) #True
#對於一般變數而言，內容相同的變數 is 和 == 的結果相同
print("==Test3==")
x=3
y=3
print(x is y) #True
print(x==y) #True
x=3
y=x
print(x is y) #True
print(x==y) #True
#對於串列list而言，內容相同的變數(==成立)，is 一定要id都相同才成立
print("==Test4==")
x=[1,2,3,4]
y=[1,2,3,4]
print(x is y) #False
print(x==y) #True
x=[1,2,3,4] 
y=x
print(x is y) #True
print(x==y) #True
```
