```python
#chap2-07a
#classes串列有八個項目
classes=['忠','孝','亻二','愛','亻言','義','禾口','平']
#印出串列
print(classes)
#印出串列項目個數(項目的編號分別是0~項目個數-1)
print(len(classes))
#印出串列項目0
print(classes[0])
#印出串列中項目內容為'孝'的項目編號
print(classes.index('孝'))
#分別將項目編號2,4,6的項目內容變更
classes[2]='仁'
classes[4]='信'
classes[6]='和'
#印出串列
print(classes)
#串列從後面增加項目"公"
classes.append("公")
print(classes)
#串列從後面增加項目"毅"
classes.append("毅")
print(classes)
#為了讓後面的順序依序為"公誠勤毅"，要將"誠""勤"依序插入
#為了在"毅"(目前的項目編號為9)前面插入"誠"
classes.insert(9,'誠')
print(classes)
#為了在"毅"(目前的項目編號為10)前面插入"勤"
classes.insert(10,'勤')
print(classes)
#移除項目內容為'勤'的項目
classes.remove('勤')
print(classes)
#移除項目標號8的項目
del classes[8]
print(classes)
#移除串列最後端的項目
classes.pop()
print(classes)
#複製到一個新的串列newclasses
newclasses=classes.copy()
newclasses.sort()
print(newclasses)
```

```python
#chap2-07b
classes=['忠', '孝', '仁', '愛', '信', '義', '和', '平']
classnum=[1,2,3,4,5]
#兩個串列合併
allclass=classes+classnum
print(allclass)
#由大到小排列
classnum.sort(reverse=True)
print(classnum)
#如果串列中有不同形態的內容(有字串和數值)是無法排列的
#allclass.sort()
```
