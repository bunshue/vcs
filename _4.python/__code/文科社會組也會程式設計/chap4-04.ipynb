```python
#chap4-04a
#i的初始值為1，終止值小於100，遞增值為1
for i in range(1,100):
  #當i>5成立時就跳到for迴圈之外
  if i>5:
    break
  print(i,end=" ")
```

```python
#chap4-04b
#i的初始值為1，終止值小於100，遞增值為1
for i in range(1,100):
  #當i為偶數時(除以2餘數為0)就跳回for迴圈起點(下面的指令都略過)
  if i%2==0:
    continue
  print(i,end=" ")
```

```python
#chap4-04c
#while True: 構成一個無窮迴圈，只有執行break指令才能跳出來
i=1
while True:
  if i>5:
    break
  print(i,end=' ')
  i=i+1
```

```python
#chap4-04d
#列出符合 x*x<240 的所有正整數
x=0
while True:
  x=x+1
  if x*x<240:
    print(x,end=" ")
  else:
    break
```

```python
#chap4-04e
#列出符合 x*x<240 的所有正奇數
x=0
while True:
  x=x+1 
  if x%2==0:
    continue
  if x*x<240:
    print(x,end=" ")
  else:
    break; 
```
