```python
#chap2-05
#等待鍵盤輸入至變數name和school
#input("提示內容")
name=input("請輸入姓名:")
school=input("請輸入就讀學校:")
#print預設會換行
print(name)
print(school)
#print列舉變數內容，變數之間預設為輸出一個空白
print(name,school)
#print列舉變數內容，改變變數之間輸出為'--'，sep為seperate分隔的意思
print(name,school,sep='--')
#改變print預設的換行為'--'，就不會換行，下一行會接著輸出
print(name,end='--')
print(school)
```
