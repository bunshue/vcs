```python
#chap4-06a
#判斷n是否為質數
#如果找到有2到math.sqrt(n)中有一個整數是n的因數，則n就不是質數
import math
n=int(input("請輸入一個整數:"))
i=2
while i<=math.sqrt(n):
  if n%i==0:
    print(n,"不是質數")
    break
  i=i+1
else:
  print(n,"是質數")
```

```python
#chap4-06b
#巢狀迴圈(外層包內層)
#列出九九乘法表
for i in range(1,10):
  print(end=" | ")
  for j in range(1,10):
    print(i,"*",j,"=","%2d" % (i*j), sep = '', end=" | ")
  print()
```

```python
#chap4-06c
#列出list列表中的所有元素
classes=['忠', '孝', '仁', '愛', '信', '義', '和', '平']
for classname in classes:
  print(classname,end=" ")
```

```python
#chap4-06d
#列出dictionary字典中所有keys和values
dict1={'dog': '狗', 'cat': '貓', 'pig': '豬', 'cow': '牛', 'monkey': '猴', 'mouse': '鼠', 'money': '猴子'}
for eng,chin in dict1.items():
  print(eng,":",chin,end=" ")
print()
for eng in dict1.keys():
  print(eng,":",dict1.get(eng),end=" ")
print()
for chin in dict1.values():
  print(chin,end=" ")
```
