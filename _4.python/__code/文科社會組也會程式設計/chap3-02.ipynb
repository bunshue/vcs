```python
#chap3-02a
#依據民法112年將成年下修為18歲，輸入年齡判斷是法為成年
#輸入後預設是字串型態存入變數
a=input("請輸入年齡:")
print(type(a))
#字串的比較
if a>="18":
  print("成年")
```

```python
#chap3-02d
#依據成績分成以下等第
#優:90<=score<=100
#甲:80<=score<90
#乙:70<=score<80
#丙:60<=score<70
#丁:score<60
score=int(input("請輸入成績:"))
if score>=90 and score<=100:
  print("優")
if score>=80 and score<90:
  print("甲")
if score>=70 and score<80:
  print("乙") 
if score>=60 and score<70:
  print("丙") 
if score<60:
  print("丁")   
```
