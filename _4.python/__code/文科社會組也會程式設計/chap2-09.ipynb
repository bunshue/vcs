```python
#chap2-09
#集合set，集合的特點是相同元素不會重複出現
#集合set1有四個項目
set1={1,2,3,4}
#集合set2將串列list轉換成集合
set2=set([1,2,3,4])
#集合set3將串列list轉換成集合
set3=set(["apple","banana","pineapple","guava"])
print(set1)
print(set2)
print(set3)
#集合set3增加一個項目
set3.add("orange")
print(set3)
#集合set3移除一個項目
set3.remove("apple")
print(set3)
#集合set4將字串轉換成集合
set4=set("apple")
print(set4)
#集合set5將字串轉換成集合
set5=set("pineapple")
print(set5)
#集合的聯集(在set4或在set5)
print(set4|set5)
#集合的交集(在set4且在set5)
print(set4&set5)
#集合的差及(在set4不在set5)
print(set4-set5)
#集合的差及(在set5不在set4)
print(set5-set4)
#愛迪達的英文有幾個不同的字母
set6=set("ADIDAS")
print(set6)
#"庭院深深深幾許，淚眼問花花不語"句子中有幾個不同的字?
seta=set("庭院深深深幾許")
setb=set("淚眼問花花不語")
print(len(seta|setb))
```
