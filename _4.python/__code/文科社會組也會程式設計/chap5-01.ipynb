```python
#chap5-01a
#bool()回傳參數的布林值
##只有0、空字串代表布林值的False，其餘都是True
print("==Test1==")
print(bool(-1))
print(bool(1))
print(bool(0))
print(bool("0"))
print(bool(""))
#int()回傳參數的整數值
print("==Test2==")
print(int("123"))
print(int(123.4))
print(int(123.5))
#float()回傳參數的浮點數值
print("==Test3==")
print(float("123.4"))
print(float(123))
#comlex()回傳參數的複數值(a+bj,a是實部,b是虛部)
print("==Test4==")
print(complex(1,2))
##複數相加，(a+bj)+(c+dj)=(a+c)+(b+d)j
print(complex(1,2)+complex(1,-2))
##複數相乘，(a+bj)*(c+dj)=(a*c-b*d)+(a*d+b*c)j
print(complex(1,2)*complex(1,-2))
```

```python
#chap5-01b
#bin()回傳參數的二進位值
##輸出結果數字前面0b代表二進位(binary)
print(bin(10))
#oct()回傳參數的八進位值
##輸出結果數字前面0o代表八進位(octal)
print(oct(10))
#hex()回傳參數的十六進位值
##輸出結果數字前面0x代表十六進位(hexdecimal)
print(hex(10))
```

```python
#chap5-01c
#abs()回傳參數的絕對值
print("==Test1==")
print(abs(-2))
#round()回傳參數的五判斷(奇進偶捨)六入值
##round(3.5)和round(4.5)可觀察到，前面結果進位(奇進)，後面結果捨去(偶捨)
##預設是取到整數位，根據小數第一位(如果是5要看個位數，奇進偶捨)判別
print("==Test2==")
print(round(3.5))
print(round(3.6))
print(round(4.5))
print(round(4.6))
##指定取到小數第一位，根據小數第二位(如果是5要看小數第一位，奇進偶捨)判別
print("==Test3==")
print(round(1.35,1))
print(round(1.36,1))
print(round(1.45,1))
print(round(1.46,1))
#pow(a,b)回傳a的b次方，如果有三個參數則求完次方後對第三個數取餘數
print("==Test4==")
print(pow(2,10))
print(pow(2,10,3))
#sum()回傳串列(list)參數的總和
print(sum([1,2,3,4]))
#min()回傳所有參數的最小值
print(min(1,2,3,4,5))
#max()回傳所有參數的最大值
print(max(1,2,3,4,5))
```

```python
#chap5-01d
#ord()回傳參數字元對應的的編碼位置
print("==Test1==")
print(ord("H"))
print(ord("你"))
print(ord("好"))
#chr()回傳參數編碼位置對應的字元
print("==Test2==")
print(chr(72))
print(chr(20320))
print(chr(22909))
##輸出'A'之後的10個英文字母
print("==Test3==")
for i in range(65,75):
  print(chr(i),end='')
print()
##輸出'你'之後的10個中文字
for i in range(20320,20330):
  print(chr(i),end='')
print()
#str()回傳參數為字串
print("==Test4==")
print(str(123)+"456")
#ascii()回傳參數的字串表達形式
##如果字串含有非ASCII字元，所有非ASCII字元會以Unicode跳脫字元的方式呈現
print(ascii("Ab123"))
print(ascii("hello你好".encode('utf-8')))
```

```python
#chap5-01e
print("==Test1==")
str1="hello你好"
##中英文混合字數(都算1個)
print(len(str1))
##中英文混合字數(中文算3個)
print(len(str1.encode('utf-8')))
##中文字數(每多一個中文，一個中文字用3bytes儲存，相差的長度就是2)
print(int((len(str1.encode('utf-8'))-len(str1))/2))
#format()回傳對齊格式後的字串
print("==Test2==")
print("12345678901234567890")
##長度20，置中對齊
print(format(str1,"^20"))
##長度20，靠左對齊
print(format(str1,"<20"))
##長度20，靠右對齊
print(format(str1,">20"))
#bytes()回傳字串轉成bytes型態，不可以修改項目內容
print("==Test3==")
bytes1=bytes(str1,encoding="utf-8")
for i in range(0,len(bytes1)):
  print(hex(bytes1[i]),end=' ')
print()
#bytearray()回傳字串轉成bytearray型態，可以修改項目內容
bytearr1=bytearray(str1,encoding='utf-8')
for i in range(0,len(bytearr1)):
  print(hex(bytearr1[i]),end=' ')
print()
##把h換成H
print("==Test4==")
bytearr1[0]=ord('H')
print(str(bytearr1,encoding='utf-8'))
##把'好'換成'你'(每個中文用了3個bytes)
bytearr1[8]=bytearr1[5]
bytearr1[9]=bytearr1[6]
bytearr1[10]=bytearr1[7]
print(str(bytearr1,encoding='utf-8'))
```

```python
#chap5-01f
print("==Test1==")
str1="hello\t你好\n我是kitty"
##中英文混合字數(都算1個)
print(len(str1))
##中英文混合字數(中文算3個)
print(len(str1.encode('utf-8')))
##中文字數(每多一個中文，一個中文字用3bytes儲存，相差的長度就是2)
print(int((len(str1.encode('utf-8'))-len(str1))/2))
print("==Test2==")
#輸出字串
print(str1)
#輸出字串的原本模樣
print(repr(str1))
#bytearray()回傳字串轉成bytearray型態，可以修改項目內容
print("==Test3==")
bytearr1=bytearray(str1,encoding='utf-8')
for i in range(0,len(bytearr1)):
  print(hex(bytearr1[i]),end=' ')
print()
print("==Test4==")
##把h換成H
bytearr1[0]=ord('H')
##把'好'換成'你'(每個中文用了3個bytes)
bytearr1[9]=bytearr1[6]
bytearr1[10]=bytearr1[7]
bytearr1[11]=bytearr1[8]
##把'\t'換成空白，Tab的ASCII碼是9
bytearr1[5]=ord(' ')
##把'\n'換成空白，LF(LineFeed)的ASCII碼是10
bytearr1[12]=ord(' ')
print(str(bytearr1,encoding='utf-8'))
```

```python
#chap5-01g
print("==Test1==")
#tuple()回傳元組型態
print(tuple([3,1,2]))
#list()回傳串列型態
print(list((1,2,3)))
#set()回傳集合型態
print(set([1,2,3]))
print("==Test2==")
list1=[3,1,2]
#len()回傳tuple或list的項目個數
print(len(list1))
print(list1)
#sorted()回傳排序後的後的串列(預設為升序)
print(sorted(list1))
#sorted()回傳排序後的後的串列(指定為降序)
print(sorted(list1,reverse=True))
print("==Test3==")
#集合a可以修改項目
a={1,2,3}
a.remove(2)
print(a)
#串列a可以修改項目
a=[1,2,3]
a.remove(2)
print(a)
#串列b是無法修改的
b=frozenset([1,2,3])
#b.remove(2)
print(b)
```

```python
#chap5-01h
#dict()回傳字典物件
print("==Test1==")
print(dict(one=1,two=2))
print(dict({'one':1,'two':2}))
print(dict(zip(('one','two'), (1,2))))
print(dict([['two',2],['one',1]]))
print("==Test2==")
#all()若所有項目轉換成Bool都是True則回傳True，否則回傳False
print(all([1,'hello',True]))
#any()若其中有一個項目轉換成Bool是True則回傳True，否則回傳False
print(any([0,False,''])) 
#zip()將物件中對應的項目打包成一個元組(tuple)，當然每個物件的項目長度要一樣，否則只會取最短的項目
list1=['one','two','three']
list2=[1,2,3]
list3=['甲','乙','丙']
##將三個list打包成一個元組tuple1
print("==Test3==")
tuple1=zip(list1,list2,list3)
for a,b,c in tuple1:
  print(a,b,c)
##將二個list打包成一個元組再轉成字典(dict)，字典只能由二個list打包
print("==Test4==")
dict1=dict(zip(list1,list2))
print(dict1)
for a,b in dict1.items():
  print(a,b)
##將二個list打包成一個元組再轉成字典(dict)，字典只能由二個list打包
print("==Test5==")
dict2=dict(zip(list3,list2))
print(dict2)
for a,b in dict2.items():
  print(a,b)
```
