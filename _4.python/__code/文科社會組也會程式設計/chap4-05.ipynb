```python
#chap4-05a
#for else 語法
#判斷是否因為break中途離開for
#正常結束沒有break就會執行else敘述
for i in range(1,5):
  if i==4:
    print("for中斷結束")
    break;
else:
  print("for正常結束")
for i in range(1,5):
  if i==5:
    print("for 中斷結束")
    break;
else:
  print("for 正常結束")
```

```python
#chap4-05b
#while else 語法
#判斷是否因為break中途離開while
#正常結束沒有break就會執行else敘述
i=1
while i<5:
  if i==4:
    print("while 中斷結束")
    break;
  i=i+1
else:
  print("while 正常結束")
i=1
while i<5:
  if i==5:
    print("while 中斷結束")
    break;
  i=i+1
else:
  print("while 正常結束")
```

```python
#chap4-05c
#不用for else 語法的替代方法，就是加一個變數
#判斷是否因為break中途離開for
breakflag=False
for i in range(1,5):
  if i==4:
    breakflag=True    
    break;
if not breakflag:
  print("for正常結束")
else:
  print("for中斷結束")

breakflag=False  
for i in range(1,5):
  if i==5:
    breakflag=True     
    break;
if not breakflag:
  print("for正常結束")
else:
  print("for中斷結束")
```

```python
#chap4-05d
#不用while else 語法的替代方法，就是加一個變數
#判斷是否因為break中途離開while
breakflag=False
i=1
while i<5:
  if i==4:
    breakflag=True    
    break;
  i=i+1
if breakflag:
  print("while 中斷結束")  
else:
  print("while 正常結束")
breakflag=False  
i=1
while i<5:
  if i==5:
    breakflag=True   
    break;
  i=i+1
if breakflag:
  print("while 中斷結束")  
else:
  print("while 正常結束")
```
