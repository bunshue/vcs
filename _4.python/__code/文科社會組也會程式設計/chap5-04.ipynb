```python
#chap5-04a
#arguments元組參數
def function1(*args):
  print(type(args))
  print("元組參數",args)

function1(1,2,3)
```

```python
#chap5-04b
#keyword arguments字典參數
def function2(**kwargs):
  print(type(kwargs))
  print("字典參數",kwargs)

function2(x=1,y=2)
```

```python
#chap5-04c
def function3(arg,*args,**kwargs):
  print("參數 ",arg)  
  print("元組參數",args)
  print("關鍵字參數",kwargs)
  print(type(arg))
  print(type(args))
  print(type(kwargs))
function3(3,1,2,3,x=1,y=2)
```

```python
#chap5-04e
def minimum(a,b):
  """使用minumum可以找出a,b較小的值
  傳入參數:
    a:第一個參數
    b:第二個參數    
  回傳結果:
    回傳a,b中較小的值
  """
  if a>b:
    return b
  else:
    return a

help(minimum)
print(minimum.__doc__)
```
