```python
#chap2-06a
#整數型態
a=5
print(type(a))
#浮點數型態
b=3.14
print(type(b))
#字串型態
c='5'
print(type(c))
#布林值型態
d=True
print(type(d))
```

```python
#chap2-06b
#數值轉布林值
print(bool(0))
print(bool(1))
#字串轉數值,10進位整數
print(int('13',10))
#字串轉數值,2進位整數
print(int('1001',2))
#字串轉數值,浮點數
print(float('3.14'))
#整數轉浮點數
print(float(1))
#數值運算
print(1+2)
#字串運算
print(str(1)+str(2))
```

```python
#chap2-06c
#a,b變數的id
a=3
b=a
print(a,b)
#a,b變數的id相同
print(id(a),id(b))
#a變數設定新值改變id
a=4
print(a,b)
print(id(a),id(b))
```

```python
#chap2-06d
#算術運算的優先順序 **(次方),*(乘)/(除)//(商數)%(餘數),+(加)-(減)
#優先順序相同時由左而右
print(3+5-4*6/2**2)
#整數除法求商數
print(21//5)
#整數除法求餘數
print(21%5)
#變數的累加
a=0
print(id(a),a)
a=a+1
print(id(a),a)
#a+=1與a=a+1相同
a+=1
print(id(a),a)
```

```python
#chap2-06e
#字串中有\的處理
str1="C:\\Windows\\System32\\mspaint.exe"
print(str1)
#字串可以用''包圍或""包圍，可視情況決定，若內容有"存在，當然用''包圍
str1='Franklin wrote, "Remember that time is money."'
print(str1)
#若內容有"存在，用\"也可以
str1="Franklin wrote, \"Remember that time is money.\""
print(str1)
#字串可以用''包圍或""包圍，可視情況決定，若內容有'存在，當然用""包圍
str1="Mary's dog is white."
print(str1)
#若內容有'存在，用\'也可以
str1='Mary\'s dog is white.'
print(str1)
#若內容有換行存在，可使用'''或"""前後包圍
str1="""Franklin wrote, \"Remember that time is money.\"
Mary's dog is white."""
print(str1)
#若內容有換行存在，用\n也可以
str1="Franklin wrote, \"Remember that time is money.\"\nMary\'s dog is white."
print(str1)
```

```python
#chap2-06f
str1="01234"
str2='56789'
#字串串接
str3=str1+str2
print(str3)
#部分字串 字串名[起始位置(預設值為0):終止位置(不包含此字元,預設值為字串長度):間隔數(預設值為1)]
#起始位置、終止位置若為負，從右邊開始數；若為正，從左邊開始數
#取第2個字(從0開始數)
print("str3[2]=",str3[2])
#第2個字開始到結束(從0開始數)
print("str3[2:]=",str3[2:])
#第2個字開始到第6個字(從0開始數)
print("str3[2:7]=",str3[2:7])
#第2個字開始到第6個字(從0開始數)，間隔2個
print("str3[2:7:2]=",str3[2:7:2])
#第0個字開始到第6個字(從0開始數)
print("str3[:7]=",str3[:7])
#第0個字開始到第-2個字(從0開始數)
print("str3[:-1]=",str3[:-1])
#第-5個字開始到第-4個字(從0開始數)
print("str3[-5:-3]=",str3[-5:-3])
#倒著數每一個字(從0開始數)
print("str3[::-1]=",str3[::-1])
#同上
print("str3[-1:-11:-1]=",str3[-1:-11:-1])
```
