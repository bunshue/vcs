```python
#chap4-02a
#ord()是將字元轉成ASCII碼
#chr()是將ASCII碼轉成字元
#ord()和chr()互為反函數
print(chr(ord('A')))
print(ord(chr(65)))
```

```python
#chap4-02b
#i的初始值為A字元的ASCII碼(65)，終止值小於Z字元的ASCII碼(90)+1，遞增值為1
for i in range(ord("A"),ord("Z")+1,1):
  print(chr(i), end = '')
```

```python
#chap4-02c
#輸入一個數n，求1+2+...+n的總和
n=int(input("求1+2+...+n,n=?"))
sum=0
#i的初始值為1，終止值小於n+1，遞增值為1
#也就是i從1數到n，用一個變數sum來累加(注意sum初始值要設為0)
for i in range(1,n+1):
  sum=sum+i
print(sum)
```
