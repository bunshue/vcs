```python
#chap4-03a
#i的初始值為1，終止值小於6，遞增值為1
i=1 #初始值
while i<6:  #終止值
  print(i,end='')
  i=i+1 #遞增值
```

```python
#chap4-03b
#i的初始值為5，終止值大於0(大於等於1)，遞減值為-1
i=5
while i>=1:
  print(i, end='')
  i=i-1
```

```python
#chap4-03c
#i的初始值為0，終止值小於等於5，遞增值為2
i=0
while i<=5:
  print(i,end='')
  i=i+2 
```

```python
#chap4-03d
#i的初始值為A字元的ASCII碼(65)，終止值小於等於Z字元的ASCII碼(90)，遞增值為1
i=ord('A')
while i<=ord('Z'):
  print(chr(i),end='')
  i=i+1  
```

```python
#chap4-03e
#輸入一個數n，求1+2+...+n的總和
n=int(input("求1+2+...+n,n=?"))
sum=0
#i的初始值為1，終止值小於等於n，遞增值為1
#也就是i從1數到n，用一個變數sum來累加(注意sum初始值要設為0)
i=1
while i<=n:
  sum=sum+i
  i=i+1
print(sum)
```
