```python
#chap2-08a
#字典dict1有三個項目(每個項目由關鍵值keys:對應值values)
dict1={"dog":"狗","cat":"貓","pig":"豬","monkey":"猴"}
#字典dict2有二個項目
dict2={"monkey":"猴","mouse":"鼠"}
#字典dict1增加一個項目('cow'是一個關鍵值)
dict1['cow']="牛"
#字典dict2更新一個項目('mokey'是一個已存在的關鍵值)
dict2['monkey']="猴子"
#分別列出字典dict1,dict2
print(dict1)
print(dict2)
#用字典dict2內容去更新dict1內容(若關鍵值不存在就是新增，關鍵值已存在就是更新)
dict1.update(dict2)
print(dict1)
print(dict1["dog"])
#取出關鍵值keys為"dog"的對應值values
print(dict1.get("dog"))
#取出關鍵值keys為"dogg"的對應值values，因為keys不存在會造成錯誤
#print(dict1["dogg"])
#取出關鍵值keys為"dogg"的對應值values
#使用get雖然keys不存在不會造成錯誤，輸出None
print(dict1.get("dogg"))
#刪除關鍵值keys為"dog"的項目
del dict1["dog"]
print(dict1)
#清除整個dict1字典
dict1.clear()
print(dict1)
```

```python
#chap2-08b
#字典dict1有三個項目
dict1={"dog":"狗","cat":"貓","pig":"豬"}
#字典dict2=dict1,此時dict2和dict1的id相同，佔用相同記憶體
dict2=dict1
#字典dict3複製dict1的內容,此時dict3和dict1的id不同，佔用不同記憶體
dict3=dict1.copy()
print(id(dict1))
print(id(dict2))
print(id(dict3))
#因為dict2和dict1的id相同，佔用相同記憶體
#修改dict2等於修改dict1
dict2["pig"]="笨豬"
print(dict1)
print(dict2)
print(dict3)
```
