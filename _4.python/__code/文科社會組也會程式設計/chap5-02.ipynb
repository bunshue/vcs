```python
#chap5-02a
#split()切割字串產生串列(list)
string1="A,B,C"
list1=string1.split(',')
print(list1)
list2=string1.split(',B,')
print(list2)
#join()指定連接字串將串列(list)項目合併字串
string2=','.join(list1)
print(string2)
string3='-'.join(list1)
print(string3)
```

```python
#chap5-02b
string1="床前明月光,疑是地上霜.舉頭望明月,低頭思故鄉."
#find()從字串左邊開始尋找第一個符合參數的部分字串
##第一個字的位置是0開始算起
print("==Test1==")
print(string1.find("月"))
##若找不到符合參數的部分字串就回傳-1
print(string1.find("日"))
#rfind()從字串右邊開始尋找第一個符合參數的部分字串
print(string1.rfind("月"))
#startswith()判斷字串是否以參數的部分字串開始
print(string1.startswith("床前"))
#startswith()判斷字串是否以參數的部分字串結束
print(string1.endswith("故鄉."))
print("==Test2==")
#replace()將字串中有第一個參數的部分字串換
string1=string1.replace(',','，')
string1=string1.replace('.','。')
print(string1)
print("==Test3==")
#count()計算字串中參數部分字串的出現個數
print(string1.count('月'))
print(string1.count('，'))
print(string1.count('。'))
```

```python
#chap5-02c
print("01234567890123456789")
print("==Test1==")
string1=str(153)
print(string1)
#center()指定寬度置中對齊
print(string1.center(20))
#ljust()指定寬度靠左對齊
print(string1.ljust(20))
#rjust()指定寬度靠右對齊
print(string1.rjust(20))
print("==Test2==")
#center()指定寬度置中對齊，指定補齊填補字元
print(string1.center(20,"-"))
#ljust()指定寬度靠左對齊，指定補齊填補字元
print(string1.ljust(20,"-"))
#rjust()指定寬度靠右對齊，指定補齊填補字元
print(string1.rjust(20,"-"))
print("==Test3==")
#zfill()指定寬度靠右對齊，以'0'補齊
print(string1.zfill(20))
```

```python
#chap5-02d
#upper()所有小寫字母轉成大寫字母
string1="This is a book."
print(string1.upper())
#lower()所有大寫字母轉成小寫字母
print(string1.lower())
#swapcase()大小寫字母互換
print(string1.swapcase())
#capitalize()第一個字母轉成大寫字母
string2="this is a book."
print(string2.capitalize())
#capitalize()每個單字的一個字母轉成大寫字母
print(string2.title())
```

```python
#chap5-02e
string1="  Hello Kitty  "
#strip()移除左右兩側的空白
print(string1.strip())
#lstrip()移除左側的空白
print(string1.lstrip())
#rstrip()移除右側的空白
print(string1.rstrip())
#strip()移除左右兩側的換行符號
string2="Hello Kitty\n"
##輸出string2的後面有一個空白行，因為有\n
print(string2)
##移除換行
print(string2.strip('\n'))
```
