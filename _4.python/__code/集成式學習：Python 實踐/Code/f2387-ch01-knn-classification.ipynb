```python
# ---第 1 部分---
# 載入函式庫與資料
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_breast_cancer
from sklearn import metrics
bc = load_breast_cancer()

# ---第 2 部分---
# 把資料分為訓練資料集與驗證資料集
train_x, train_y = bc.data[:400], bc.target[:400]
test_x, test_y = bc.data[400:], bc.target[400:]

# ---第3 部分---
# 初始化、訓練並評估模型
knn = KNeighborsClassifier()
knn.fit(train_x, train_y)
acc = metrics.accuracy_score(test_y, knn.predict(test_x))

# ---第 4部分---
# 顯示模型訓練結果
print('---KNN on breast cancer dataset---')
print('Accuracy: %.2f \n'%acc)
print(metrics.confusion_matrix(test_y, knn.predict(test_x)))
```
