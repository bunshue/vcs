```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

data = pd.read_csv('/kaggle/input/btcusd/BTC-USD.csv')
data = data.dropna()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.set_index('Date', drop=True, inplace=True)
```

```python
plt.figure(figsize = (8, 8))
data.Close.plot()
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bitcoin')
plt.xticks(rotation = 70)
plt.show()
```

```python
# 計算自相關
fig, axes = plt.subplots(figsize =(8, 8))
plot_acf(data.Close, axes, lags = 30)
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.show()
```

```python
# 計算變化百分比
data['diffs'] = (data.Close.diff() / 
                 data.Close.shift(periods=1, 
                                  fill_value=0)).values
data['diffs'] = data['diffs'].fillna(0)
```

```python
# 繪製圖形
plt.figure(figsize = (8, 8))
data.diffs.plot()
plt.xlabel('Date')
plt.ylabel('Change %')
plt.title('Transformed Data')
plt.xticks(rotation = 70)
plt.show()

```

```python
fig, axes = plt.subplots(figsize =(8, 8))
plot_acf(data.diffs, axes, lags=60)
plt.xlabel('Date')
plt.ylabel('Correlation')
plt.show()
```

```python
plt.figure(figsize = (8, 8))
data.diffs.rolling(30).std().plot()
plt.xlabel('Date')
plt.ylabel('Std. Dev.')
plt.title('Transformed Data Rolling Std.Dev.')
plt.xticks(rotation = 70)
plt.show()
```
